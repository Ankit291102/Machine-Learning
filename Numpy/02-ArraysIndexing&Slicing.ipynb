Array Indexing

```python
import numpy as np
```

NUMPY INDEXING AND SLICING OF VECTORS

```python
arr=np.arange(11,21)
arr
```

```python
arr[5]
```

```python
arr[1:5]
```

```python
arr[:5]
```

```python
arr[:]
```

```python
arr[3:]
```

```python
arr[3::2]
```

NUMPY INDEXING AND SLICING OF MATRIX

```python
arr=np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
arr[1]
```

```python
arr[0][0]
```

```python
arr[0,0]
```

```python
arr[0,4]
```

```python
arr[5,4]
```
