Finding Missing Data

```python
import numpy as np 
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Removing Missing Data -- Works in row wise

```python
df.dropna() # removes rows with any NaN values
```

```python
df.dropna(thresh=3) # removes rows with less than 3 non-NaN values  
```

Filling the missing data 

```python
df.fillna(0)
```

```python
values = {
    'A': 0,
    'B': 200,
    'C': 300,
    'D': 400
}

df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
