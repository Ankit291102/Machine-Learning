```python
import numpy as np
import pandas as pd
```

Creating a DataFrame 

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
df
```

```python
new_data = [
    ['John', 'Anna', 'Peter', 'Linda'],
    [28, 34, 29, 42],
    ['New York', 'Paris', 'Berlin', 'London'],
    [65000, 70000, 62000, 85000]
]
df1 = pd.DataFrame(new_data)
df1
```

Selecting rows based on conditions:

```python
Raw_data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(Raw_data)
df
```

```python
df.loc[0]
```

```python
df.loc[[0,1]]
```

```python
df.iloc[3]
```

Selecting Subsets of Rows and Columns

```python
df.loc[[0,1]][['City','Salary']]
```

```python
df.loc[0:2][['City','Salary']]
```

Conditional selection in pandas allows you to filter data based on specific conditions. You can use boolean indexing to select rows that meet certain criteria. Here's an example:

```python
df[df['Age']>30]
```

```python
df[(df['Age']>30) & (df['City']=='Paris')]
```
