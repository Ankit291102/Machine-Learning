```python
import numpy as np
import pandas as pd
```

GroupBy

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#Grouping by 'Category' and aggregating 'Sales' and 'Quantity' using sum 
df.groupby('Category').aggregate({
    'Sales': 'sum',
    'Quantity': 'sum'
})
```

```python
df.groupby('Category')['Sales'].sum()
```

```python
df.groupby(['Category','Store'])['Sales'].sum()
```

Aggregation using multiple functions for different columns

```python
df['Sales'].mean(),df['Sales'].std(),df['Sales'].mode()
```

```python
df['Sales'].agg(['sum', 'mean', 'max', 'min','count','std','var','median'])
```
