```python
import pandas as pd
import numpy as np
```

Pivot Tables are used to summarize and aggregate data in a tabular format. They allow you to transform and reshape your data, making it easier to analyze and visualize. In this notebook, we will explore how to create pivot tables using the Pandas library in Python.

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': [
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South'
    ],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': [
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice'
    ]
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values='Sales', index='Region', columns='Product', aggfunc='sum', fill_value=0) 
```

```python
pd.pivot_table(df,values=['Sales','Units'], index='Region', columns='Product', aggfunc='sum', fill_value=0)
```

Cross Tabs

```python
pd.crosstab(df['Region'], df['Product'])
```

```python
pd.crosstab(df["Region"], df["Product"], margins=True)
```

```python
pd.crosstab(
    df["Region"],
    df["Product"],
    normalize=True
)
```

```python
pd.crosstab(
    df["Region"],
    df["Product"],
    normalize="index"
)
```

```python
pd.crosstab(
    df["Region"],
    df["Product"],
    normalize="columns"
)
```
