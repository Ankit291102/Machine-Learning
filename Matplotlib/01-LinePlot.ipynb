```python
# From a pandas dataframe
df=pd.read_csv('sharma-kohli.csv')
df.head()
```

```python
df.plot(x='index',y='RG Sharma',kind='line')
```

```python
plt.plot(df['index'], df['RG Sharma'])
```

```python
# Plotting multiple plots
plt.plot(df['index'], df['RG Sharma'], label='RG Sharma')
plt.plot(df['index'], df['V Kohli'], label='V Kohli')
```

```python
# Label titles
plt.plot(df['index'], df['RG Sharma'], label='RG Sharma')
plt.plot(df['index'], df['V Kohli'], label='V Kohli')
plt.title('Runs scored by RG Sharma and V Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# Colors(hex) and line(width and style) and marker(size)
plt.plot(df['index'], df['RG Sharma'], label='RG Sharma',color='Black',linestyle='dashed',linewidth=3,marker='o',markersize=5)
plt.plot(df['index'], df['V Kohli'], label='V Kohli',color='Red',linestyle='dotted',linewidth=1,marker='s',markersize=6)
plt.title('Runs scored by RG Sharma and V Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend(loc='best')
```

```python
# Limiting the axis
price=[48000,54000,57000,49000,47000,45000,4500000]
year=[2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
```

```python
# Limiting the axis
price=[48000,54000,57000,49000,47000,45000,4500000]
year=[2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(40000,60000)
plt.xlim(2017,2020)
```

```python
# Gridlines
plt.plot(df['index'], df['RG Sharma'], label='RG Sharma',color='Black',linestyle='dashed',linewidth=3,marker='o',markersize=5)
plt.grid(True)
```

```python
# Showing the plot
plt.show()
```
