```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Scatter Plots
* Bivariate Analysis
* Use case - Finding correlation

```python
# Plot scatter plot of two variables
x=np.linspace(-10,10,50)
y=10*x+3+np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df=pd.read_csv('batter.csv')
df=df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='*')
plt.title('Scatter plot of avg vs strike_rate')
plt.xlabel('avg')
plt.ylabel('strike_rate')
```

```python
# size
tips=sns.load_dataset('tips')
tips.head()
```

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*10)
```

```python
# scatter plot using plt.plot -- Faster than plt.scatter
plt.plot(tips['total_bill'],tips['tip'],'o')
```
