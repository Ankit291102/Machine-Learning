## Handling Imbalanced Dataset

1. Up Sampling

```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_classification

x,y = make_classification(
    n_samples=1000,
    weights=[0.95,0.05],
    random_state=42,
    flip_y=0,          # No label noise
)

df=pd.DataFrame(x)
df['Target']=y
df['Target'].value_counts()
```

```python
majority=df[df['Target']==0]
minority=df[df['Target']==1]
```

```python
from sklearn.utils import resample

minority_upsampled=resample(
    minority,
    replace=True,
    n_samples=len(majority),
    random_state=42
)

balanced=pd.concat([majority,minority_upsampled])

print(balanced['Target'].value_counts())
```

##### 2. Downsampling

```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_classification

x,y = make_classification(
    n_samples=1000,
    weights=[0.95,0.05],
    random_state=42,
    flip_y=0,          # No label noise
)

df=pd.DataFrame(x)
df['Target']=y
df['Target'].value_counts()
```

```python
majority=df[df['Target']==0]
minority=df[df['Target']==1]
```

```python
from sklearn.utils import resample

majority_downsampled=resample(
    majority,
    replace=False,
    n_samples=len(minority),
    random_state=42
)

balanced=pd.concat([majority_downsampled,minority])

print(balanced['Target'].value_counts())
```
